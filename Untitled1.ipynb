```python
import sys
import os
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression, Perceptron
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score,train_test_split
from sklearn.metrics import precision_recall_fscore_support as score, accuracy_score, classification_report
from sklearn.model_selection import KFold
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import seaborn as sns
```

```python
testData = pd.read_csv("./UNSW_NB15_training-set.csv")
trainData = pd.read_csv("./UNSW_NB15_testing-set.csv")

dataset = pd.concat([testData, trainData])
dataset.drop(columns=['id'], inplace=True)
```

```python
X = dataset[dataset.columns[:-2]].to_numpy()
y = dataset['label'].to_numpy()
```

```python
#splitting dataset train : %64, test: %20 ve cross validation: %16
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
X_train, X_cv, y_train, y_cv = train_test_split(X_train, y_train, test_size=0.20, random_state=42)
```

```python
#to decrease correlation in dataset 

pd_x = pd.DataFrame(X_train).corr()
maskedData = ~(pd_x.mask(np.eye(len(pd_x), dtype=bool)).abs() > 0.7).any()
train = pd.DataFrame(X_train).loc[:,maskedData]
test = pd.DataFrame(X_test).loc[:,maskedData]
```

```python
#Normalization of data
#PCA DENENEBİLİR STANDART SCALAR YERINE

# pca = PCA(n_components=20)
# pca.fit(X)
# x_train_2=pca.transform(X)
# x_test_2=pca.transform(X_test)

sc = StandardScaler()
sc.fit(train)
X_new_train = sc.transform(train)
X_new_test = sc.transform(test)
```

```python
plt.figure(figsize=(16,12))

ax1=plt.subplot(2, 2, 1)
pd_original_X = pd.DataFrame(X_train).corr()
ax1.matshow(pd_original_X)
ax1.set_title('Original Data Correlation Matrix')


ax2 = plt.subplot(2, 2, 2)
pd_normalized_X = pd.DataFrame(X_new_train).corr()
ax2.matshow(pd_normalized_X)
ax2.set_title('Normalize Data Correlation Matrix')

```

```python
#Logistic Regression

linear_classifier = LogisticRegression(random_state = 1)
linear_classifier.fit(X_new_train, y_train)
log_reg_score = linear_classifier.score(X_new_test, y_test)
print("Logistic Regression Score: {}".format(log_reg_score))
```

```python
#Prediction ve Base Table

predicts = linear_classifier.predict(X_new_test)
y_test_flat =  y_test.flatten()
table_predict = pd.DataFrame(np.array([[y_test_flat[i], predicts[i]] for i in range(len(y_test))]), columns=['base', 'predicts'])
table_predict
```
